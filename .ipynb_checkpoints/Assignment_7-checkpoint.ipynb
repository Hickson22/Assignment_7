## Clustering

Complete the class to implement the K-means algorithm.

## Function "fit()" 
Takes in d-dimensional data and applies the EM step to find and print the cluser centroids.

Initialize your means with random samples.

## Function "predict()"
Takes in a set of observations and predicts their labels based on the nearest cluster centroids.

Returns a list of predicted labels

```python
class K_Means:
    def __init__(self, k=2, tol=0.001, max_iter=300):
        self.k = k
        self.tol = tol
        self.max_iter = max_iter

    def fit(self,data):

        self.centroids = {}

        for i in range(self.k):
            self.centroids[i] = data[i]

        for i in range(self.max_iter):
            self.classifications = {}

            for i in range(self.k):
                self.classifications[i] = []

            for featureset in data:
                distances = [np.linalg.norm(featureset-self.centroids[centroid]) for centroid in self.centroids]
                classification = distances.index(min(distances))
                self.classifications[classification].append(featureset)

            prev_centroids = dict(self.centroids)

            for classification in self.classifications:
                self.centroids[classification] = np.average(self.classifications[classification],axis=0)

            optimized = True

            for c in self.centroids:
                original_centroid = prev_centroids[c]
                current_centroid = self.centroids[c]
                if np.sum((current_centroid-original_centroid)/original_centroid*100.0) > self.tol:
                    print(np.sum((current_centroid-original_centroid)/original_centroid*100.0))
                    optimized = False

            if optimized:
                break

    def predict(self,data):
        distances = [np.linalg.norm(data-self.centroids[centroid]) for centroid in self.centroids]
        classification = distances.index(min(distances))
        return classification
```

```python
#The DataSet has Two Dimensions so we choose MinPoints = 4

#Tuning Eps / Finding the optimal value for Eps

from sklearn.neighbors import NearestNeighbors # importing the library
neighb = NearestNeighbors(n_neighbors=2) # creating an object of the NearestNeighbors class
nbrs=neighb.fit(circles_data) # fitting the data to the object
distances,indices=nbrs.kneighbors(circles_data) # finding the nearest neighbours

# Sort and plot the distances results
distances = np.sort(distances, axis = 0) # sorting the distances
#display(distances)
distances = distances[:, 1] # taking the second column of the sorted distances
plt.plot(distances) # plotting the distances
plt.show() # showing the plot

#From the Graph it can be seen that the max curviture occurs at roughly 0.20

from sklearn.cluster import DBSCAN
# cluster the data into five clusters
dbscan = DBSCAN(eps = 0.2, min_samples = 4).fit(circles_data) # fitting the model
labels_db = dbscan.labels_ # getting the labels
# Plot the clusters
plt.scatter(circles_data[:, 0], circles_data[:,1], c = labels_db, cmap= "plasma") # plotting the clusters
plt.show() # showing the plot
```

## Discuss the pros and cons for KMeans and DBSCAN.

DBSCAN: 
Pros: 1) Works well to identify arbitrarily shaped clusters. 2) can identify outliers which could potentially be the primary    goal of the exercise.  
Cons: 1) It does not work well when there is varying density across the data since Epsilon and MinPts are fixed.

KMeans:
Pros:1) It is simple, highly flexible, and efficient. 2) The efficiency of k-means implies that the algorithm is good at segmenting a dataset. 
Cons: 1) It does not allow to develop the most optimal set of clusters and the number of clusters must be decided before the analysis. 2)The order of the data has an impact on the final results.

## Using the k-means algorith find the clusters and reprodue the image using only 4 colours (k=4). 

The result looks something like the image below.

```python
display(Image(filename="src/penguin_compressed.png"))
```

```python
# Load image
image = io.imread("src/penguin.png")
# Convert from rgba to rgb
image = color.rgba2rgb(image)
# Show image dimension
print(image.shape)
```

```python
# Dimensions of original image
rows = image.shape[0]
cols = image.shape[1]

# Flatter image
image = image.reshape(rows * cols, 3)

# Form k clusters by implementing k-means clustering 
km = KMeans(n_clusters = 4)
km.fit(image)

# Replace each pixel value with nearby centroid
compressed_image = km.cluster_centers_[km.labels_]

# Reshape image to original dimension
compressed_image = compressed_image.reshape(rows, cols, 3)

# Display output image
plt.imshow(compressed_image)
```

## Print the number of clusters required for a perfect reproduction.

```python
K = range(1,20)
inertias = []

for k in K:

    model_im = KMeans(n_clusters = k)
    model_im.fit(image)
    inertias.append(model_im.inertia_)
    

plt.plot(K, inertias)
plt.xlabel('Number of clusters, k')
plt.ylabel('Inertia')
plt.show()

print("3 Clusters are required for perfect reproduction") #Not sure if we can have 2.6 clusters (ie. float)
```

## Non-negative matrix factorization
Let's apply NMF, using the tf-idf word-frequency array of Wikipedia articles. 

First explore the data. 

The data consist of a list of article titles and a list of words. The tf-idf word frequency of each word is given in a csr matrix called articles.

```python
df_articles = pd.read_csv('src/Wikipedia articles/wikipedia-vectors.csv', index_col=0)

# articles in csr matrix format
articles = csr_matrix(df_articles.transpose())
# article titles
titles = list(df_articles.columns)
# article keywords
words = pd.read_csv('src/Wikipedia articles/wikipedia-vocabulary-utf8.txt', header=None).iloc[:, 0].tolist()

# For visualization
# Use raw (non csr matrix)
raw_articles = df_articles.transpose()
df = pd.DataFrame(raw_articles, index=titles)
df.columns = words
df.head()
```

## Now fit and transform the articles. First, create an NMF instance using 6 components.

```python
model_nmf = NMF(n_components = 6)
```

## Now fit the model to articles and transform to obtain the nmf_features.


```python
nmf_features = model_nmf.fit_transform(articles)
```

## Put the features in a pandas dataframe using the article titles as index.

```python
df_nmf_features = pd.DataFrame(nmf_features,index=titles)

df_nmf_features.head()

```

## Now that you have the features, display the 6 components for article with title "Zlatan Ibrahimovi". 

```python
display(df_nmf_features.loc['Zlatan Ibrahimovi'])

```

## Now display the 6 components for article with title "Neymar". 

```python
display(df_nmf_features.loc['Neymar'])
```

What do you observe? Explain what this mean.

Answer:

Verify your assumption above by identifying the corresponding NMF components (i.e. the words that both articles have in common)

```python
from sklearn.preprocessing import normalize
norm_features = normalize(nmf_features)
df = pd.DataFrame(norm_features,index=titles)
ney_article = df.loc['Neymar']
zla_article = df.loc['Zlatan Ibrahimovi']


sim_ney = df.dot(ney_article)
sim_zla = df.dot(zla_article)


print(sim_ney.nlargest())

print(sim_zla.nlargest())

print("Corresponding NMF components are: Cristiano Ronaldo & Franck Ribry")
```
